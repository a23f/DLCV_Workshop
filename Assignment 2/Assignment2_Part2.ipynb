```python
import torch
from PIL import Image
from torchvision import transforms
#Loading Model
model = torch.load('/content/drive/MyDrive/Colab_Notebooks/entire_model.pt',map_location ='cpu')
classes=('durian','pumpkin', 'tomato','watermelon')

image_transforms=transforms.Compose([
    transforms.Resize((256,256)),
    transforms.ToTensor(),
    transforms.Normalize([0.485, 0.456, 0.406],
                         [0.229, 0.224, 0.225])
])
def classify(model, image_transforms, image_path, classes):
    model = model.eval()
    image = Image.open(image_path)
    image = image_transforms(image).float()
    image = image.unsqueeze(0)
    output= model(image)
    _, predicted = torch.max(output.data,1)
    print(classes[predicted.item()])
model.to('cpu')

classify(model, image_transforms, '/content/drive/MyDrive/Colab_Notebooks/la-fg-col1-malaysia-durians-china-pictures-003.jpg', classes)
classify(model, image_transforms, '/content/drive/MyDrive/Colab_Notebooks/images.jpg', classes)
classify(model, image_transforms, '/content/drive/MyDrive/Colab_Notebooks/giant-melon1.jpg', classes)
classify(model, image_transforms, '/content/drive/MyDrive/Colab_Notebooks/download.jpg', classes)
```
