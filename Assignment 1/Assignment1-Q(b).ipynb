```python
#path of example directory
path = "/content/drive/MyDrive/Assignments_DLCV/"
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

face_cascade = cv2.CascadeClassifier(cv2.samples.findFile(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml'))

img1 = cv2.imread(path+'Numb.jpg')

color_conv1 = cv2.cvtColor(img1, cv2.COLOR_BGR2RGB)
faces = face_cascade.detectMultiScale(color_conv1)
for (x,y,w,h) in faces:
  cv2.rectangle(color_conv1,(x,y),(x+w,y+h),(255,0,0),2)
  face = color_conv1[y:y+h, x:x+w]
  face = cv2.GaussianBlur(face,(23, 23), 30)
  color_conv1[y:y+face.shape[0], x:x+face.shape[1]] = face

img2 = cv2.imread(path+'Smile.jpg')

color_conv2 = cv2.cvtColor(img2, cv2.COLOR_BGR2RGB)
faces = face_cascade.detectMultiScale(color_conv2)
for (x,y,w,h) in faces:
  cv2.rectangle(color_conv2,(x,y),(x+w,y+h),(255,0,0),2)
  face = color_conv2[y:y+h, x:x+w]
  face = cv2.GaussianBlur(face,(23, 23), 30)
  color_conv2[y:y+face.shape[0], x:x+face.shape[1]] = face

img3 = cv2.imread(path+'Sad.jpg')

color_conv3 = cv2.cvtColor(img3, cv2.COLOR_BGR2RGB)
faces = face_cascade.detectMultiScale(color_conv3)
for (x,y,w,h) in faces:
  cv2.rectangle(color_conv3,(x,y),(x+w,y+h),(255,0,0),2)
  face = color_conv3[y:y+h, x:x+w]
  face = cv2.GaussianBlur(face,(23, 23), 30)
  color_conv3[y:y+face.shape[0], x:x+face.shape[1]] = face

plt.figure(figsize=[18,20])
plt.subplot(311);plt.imshow(color_conv1);plt.title("Numb Face Blurred")
plt.subplot(312);plt.imshow(color_conv2);plt.title("Smile Face Blurred")
plt.subplot(313);plt.imshow(color_conv3);plt.title("Sad Face Blurred")
```
